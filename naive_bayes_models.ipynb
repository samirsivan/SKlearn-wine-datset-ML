# Problem Statement:
### predict the class of wine from sklearn wine dataset and compare whose gonna perform better MultinomialNB or GaussianNB

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB, MultinomialNB
import warnings
warnings.filterwarnings('ignore')
```

```python
wine = datasets.load_wine()
```

```python
data = pd.DataFrame(wine.data, columns=wine.feature_names)
```

```python
data['target'] = wine.target
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
def show_distplot(df, row, col):
    plt.figure(figsize=(14,14))
    for i, columns in enumerate(df.columns):
        plt.subplot(row, col, i+1)
        sns.distplot(df[columns])
    plt.show()
```

### Lets see the distribution of independent features

```python
show_distplot(data.iloc[:,:-1],5,3)
```

#### from the plot, as almost all the features are quite normally distributed thus my ans would be Gaussian would perform better than multinomial

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
y = np.reshape(y, (-1,1))
X.shape, y.shape
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
```

#### For MultinomialNB it cant handle neg values so we choose minmaxscaler instead of standardscaler

```python

model = Pipeline(steps=[
    ("scale", MinMaxScaler()),
    ("multinomialNB", MultinomialNB())
])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```

#### we can choose standardscaler for GaussianNB as it can handle neg values

```python
model_1 = Pipeline(steps=[
    ("scale", StandardScaler()),
    ("multinomialNB", GaussianNB())
])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


model_1.fit(X_train, y_train)

y_pred = model_1.predict(X_test)

acc = accuracy_score(y_test, y_pred)
acc
```

### Result:
GaussianNB performed better than MultinomialNB
